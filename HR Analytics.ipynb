# Data Type Fixing & Basic Cleaning 

Purpose

Ensure dates and numeric fields behave correctly.

Business Meaning

Incorrect data types can lead to wrong attrition rates, tenure, and performance insights.

```python
# Convert date columns to datetime
date_cols = [
    "DOB",
    "DateofHire",
    "DateofTermination",
    "LastPerformanceReview_Date"
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
df.dtypes
```

```python
# Check missing values
df.isna().sum().sort_values(ascending=False)
```

“Missing values in termination and hire dates were treated as meaningful indicators rather than errors. Active employees naturally lack termination dates, and age/tenure calculations were restricted only to valid records to avoid distortion.”

```python
df.duplicated().sum()
```

```python
df["ManagerID"] = df["ManagerID"].fillna("Unknown")
```

# Feature Engineering (Critical Business Logic)

Purpose

Create meaningful business variables.

Business Meaning

Raw HR data is not enough — derived metrics drive insights.

```python
# Create Attrition Flag (Target Variable)
df["Attrition"] = df["Termd"].apply(lambda x: 1 if x == 1 else 0)
```

```python
# Calculate Age
df["Age"] = (pd.Timestamp("today") - df["DOB"]).dt.days // 365
```

```python
# Calculate Tenure (in years)
df["Tenure_Years"] = np.where(
    df["DateofTermination"].notna(),
    (df["DateofTermination"] - df["DateofHire"]).dt.days / 365,
    (pd.Timestamp("today") - df["DateofHire"]).dt.days / 365
)
```

```python
df.isnull().sum()
```

“Age was derived from date of birth. Records without DOB naturally result in missing age values, which were intentionally preserved to avoid introducing artificial demographic data.”

# Checking Outliers for Understanding

```python
def iqr_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    return lower_bound, upper_bound
```

```python
lb_salary, ub_salary = iqr_outliers(df["Salary"])

df_salary_outliers = df[
    (df["Salary"] < lb_salary) | (df["Salary"] > ub_salary)
]

df_salary_outliers[["Employee_Name", "Position", "Department", "Salary"]].head()
```

```python
lb_abs, ub_abs = iqr_outliers(df["Absences"])

df_abs_outliers = df[
    (df["Absences"] < lb_abs) | (df["Absences"] > ub_abs)
]

df_abs_outliers[["Employee_Name", "Absences", "PerformanceScore", "Attrition"]].head()
```

```python
lb_late, ub_late = iqr_outliers(df["DaysLateLast30"])

df_late_outliers = df[
    (df["DaysLateLast30"] < lb_late) | (df["DaysLateLast30"] > ub_late)
]

df_late_outliers[["Employee_Name", "DaysLateLast30", "Attrition"]].head()
```

```python
lb_eng, ub_eng = iqr_outliers(df["EngagementSurvey"])

df_eng_outliers = df[
    (df["EngagementSurvey"] < lb_eng) | (df["EngagementSurvey"] > ub_eng)
]

df_eng_outliers[["Employee_Name", "EngagementSurvey", "PerformanceScore"]].head()
```

```python
outlier_summary = {
    "Salary": len(df_salary_outliers),
    "Absences": len(df_abs_outliers),
    "DaysLateLast30": len(df_late_outliers),
    "EngagementSurvey": len(df_eng_outliers)
}

outlier_summary
```
